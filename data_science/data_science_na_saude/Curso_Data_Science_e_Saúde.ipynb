##Fonte de John Hopkings

```
def latest_by_country(data):
  return data.groupby('Country/Region').sum().iloc[:,-1]

display(latest_by_country(confirmed).head())
display(latest_by_country(deaths).head())
display(latest_by_country(recovered).head())
```

```
informations = [latest_by_country(confirmed), latest_by_country(deaths), latest_by_country(recovered)]
combined = pd.concat(informations, axis=1)
combined.columns = ['Confirmed', 'Deaths', 'Recovered']
combined
```

##Taxa de Letalidade

A taxa de letalidade é expressa em porcentagem.
É o número de casos letais dentro dos casos que foram letais e não letais.

Imagina que exista uma doença em janeiro de 2020 que afetou 100 pessoas e 5 dessas pessoas morreram (por exemplo). A letalidade foi de 5%.
E a doença deixou de existir pela história da humanidade.

Mas se a doença continuar existindo a taxa pode ser outra. Ela muda pelo passar do tempo pois encontramos protocolos de tratamento, anti-virais, vacinas, etc... E com isso a taxa muda com o passar do tempo.

Conclusão é muito difícil falar de letalidade, tentanto ser preciso, durante uma epidemia.

No começo é comum que sejam reportados apenas casos graves sejam reportados.


Uma tentativa simples de obter a taxa de letalidade é fazer mortes/confirmados

```
sum_up = combined.sum()
letality_rate_1 = sum_up['Deaths']/sum_up['Confirmed']
print(f'A taxa de letalidade 1 é: {(letality_rate_1 * 100).round(2)}%')
```

Nesse conjunto de dados do número de pessoas que pegaram covid até hoje, confirmados, 2.21% é o número de mortes de acordo com esses dados, porém, várias dessas pessoas ainda não se recuperaram, então, infelizmente, várias podem morrer, então essa taxa pode ainda piorar. Por outro lado a taxa pode diminuir pois a medida que testamos mais gente, podemos descobrir que mais pessoas possuem uma forma mais fraca do vírus com o vírus não afetando tanto elas, elas não reportam e não vão pro hospital, então, nesse caso, a taxa de letalidade pode ser muito menor.

Percebe como um resumo (um número) pode mentir? Por que se falamos apenas a taxa de letalidade de 2.21% é uma coisa, mas essa pode variar bastante de acordo com diversos fatores. Por isso precisamos sempre explicar as restrições dos dados e informações que possuímos. Resumos são difíceis mas são importantes.

#Aula 4 - Histogramas e resumindo informações

##A dificuldade de resumir informações.

Agora que criamos a letalidade total, queremos ver a letalidade por país.

```
letality_rate_1 = combined['Deaths'] / combined['Confirmed'] * 100
combined['letality_rate_1'] = letality_rate_1
combined.head()
```

Mas temos uma coluna de casos recuperados, ou seja, se temos casos recuperados e mortes, temos todos para quem acabou a conversa sobre coronavirus (sem considerar casos de reinfecção). Ou seja se 1 pessoa morreu e 5 se recuperaram a taxa de letalidade é de 1 e 6. Fazendo uma analogia do que foi feito acima:

```
sum_up = combined.sum()
letality_rate_1 = sum_up['Deaths']/sum_up['Confirmed']
letality_rate_2 = sum_up['Deaths'] / (sum_up['Recovered']+sum_up['Deaths'])
print(f'A taxa de letalidade 1 é: {(letality_rate_1 * 100).round(2)}%')
print(f'A taxa de letalidade 2 é: {(letality_rate_2 * 100).round(2)}%')
```

Por que essa diferença? O que deve ocorrer é que o tempo de recuperação deve ser longo, então como o número de casos cresce de modo exponencial, muitas pessoas não tiveram o tempo de se recuperar ou de falecer, por isso a taxa de letalidade será mais alta, então se fazemos a análise só para quem já se recuperou e não podemos dizer nada sobre os casos novos que estão entrando a taxa de letalidade vai ser realmente mais alta, pois temos um número muito grande de novos casos que não levamos em consideração, portanto, é uma análise mais pessimista.

Portanto, repare que não tem um certo ou errado entre as duas taxas, no primeiro consideramos todo mundo que está doente hoje vai se recuperar, e no segundo descartamos que todo mundo que está doente vai se recuperar.

```
letality_rate_2 = combined['Deaths'] / (combined['Recovered'] + combined['Deaths']) *100 
combined['letality_rate_2'] = letality_rate_2
combined.head()
```
